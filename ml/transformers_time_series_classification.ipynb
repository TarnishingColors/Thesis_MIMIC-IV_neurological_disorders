```python
# Oversampling

import numpy as np
from sklearn.utils import resample

# Augment the dataset to address class imbalance
def augment_data(X, y):
    # Identify the class distribution
    class_counts = Counter(y)
    max_count = max(class_counts.values())

    X_augmented, y_augmented = [], []

    for label in class_counts.keys():
        # Get all samples of the current class
        X_class = X[y == label]
        y_class = y[y == label]

        # Random oversampling to balance the classes
        if len(X_class) < max_count:
            X_resampled, y_resampled = resample(
                X_class,
                y_class,
                replace=True,
                n_samples=max_count - len(X_class),
                random_state=42
            )
            X_augmented.append(X_resampled)
            y_augmented.append(y_resampled)

    # Concatenate original and augmented data
    X_balanced = np.vstack((X, *X_augmented))
    y_balanced = np.hstack((y, *y_augmented))

    return X_balanced, y_balanced


# Example usage:
X_augmented, y_augmented = augment_data(X_train, y_train)

# Check the new class distribution
print("Class distribution after augmentation:", Counter(y_augmented))
```

```python
import torch
from torch import nn
import torch.nn.functional as F
from torch.utils.data import DataLoader, Dataset
from transformers import BertConfig, BertModel


class TimeSeriesDataset(Dataset):
    def __init__(self, sequences, targets):
        """
        sequences: numpy array of shape (num_samples, time_steps, num_features)
        targets: numpy array of shape (num_samples,)
        """
        self.sequences = torch.tensor(sequences, dtype=torch.float32)
        self.targets = torch.tensor(targets, dtype=torch.long)  # For classification tasks, targets must be of type long

    def __len__(self):
        return len(self.sequences)

    def __getitem__(self, idx):
        return self.sequences[idx], self.targets[idx]


class TransformerClassifier(nn.Module):
    def __init__(self, input_dim, seq_len, num_classes, num_hidden_size, num_attention_heads):
        super(TransformerClassifier, self).__init__()
        self.input_dim = input_dim
        self.seq_len = seq_len
        hidden_size = 128
        self.embedding_layer = nn.Linear(input_dim, hidden_size)
        config = BertConfig(
            hidden_size=hidden_size,
            num_hidden_layers=num_hidden_size,
            num_attention_heads=num_attention_heads,
            intermediate_size=256,
            max_position_embeddings=seq_len,
#             hidden_dropout_prob=0.3,
#             attention_probs_dropout_prob=0.3
            
        )
        self.transformer = BertModel(config)
        
        self.classifier = nn.Sequential(
            nn.Linear(hidden_size, hidden_size // 2),
            nn.ReLU(),
            nn.Dropout(0.3),  # Dropout after intermediate layer
            nn.Linear(hidden_size // 2, num_classes)
        )
        # batchnorm try
        

    def forward(self, x):
        x = self.embedding_layer(x)

        transformer_out = self.transformer(inputs_embeds=x).last_hidden_state

        logits = self.classifier(transformer_out[:, 0, :])
        return F.softmax(logits, dim=1)
```

```python
def train_model(model, train_dataloader, val_dataloader, epochs, patience, lr):
    optimizer = torch.optim.Adam(model.parameters(), lr=lr)
    criterion = nn.CrossEntropyLoss()
    model.to(device)
    best_loss = float('inf')
    patience_counter = 0

    for epoch in range(epochs):
        model.train()
        total_loss = 0
        for batch in train_dataloader:
            inputs, targets = batch
            inputs, targets = inputs.to(device), targets.to(device)

            optimizer.zero_grad()
            outputs = model(inputs)

            # Calculate the training loss
            loss = criterion(outputs, targets)
            loss.backward()
            optimizer.step()

            total_loss += loss.item()
        
        avg_train_loss = total_loss / len(train_dataloader)
#         print(f"Epoch {epoch + 1}, Training Loss: {avg_train_loss:.4f}")

        model.eval()
        val_loss = 0
        with torch.no_grad():
            for batch in val_dataloader:
                inputs, targets = batch
                inputs, targets = inputs.to(device), targets.to(device)

                outputs = model(inputs)

                loss = criterion(outputs, targets)
                val_loss += loss.item()
        
        avg_val_loss = val_loss / len(val_dataloader)
        print(f"Epoch {epoch + 1}, Training Loss: {avg_train_loss:.4f}, Validation Loss: {avg_val_loss:.4f}")

        # Early stopping logic
        if avg_val_loss < best_loss:
            best_loss = avg_val_loss
            patience_counter = 0  # Reset patience counter if improvement
#             print(f"Validation loss improved to {best_loss:.4f}")
        else:
            patience_counter += 1
#             print(f"No improvement for {patience_counter} epoch(s).")

        if patience_counter >= patience:
#             print(f"Early stopping triggered after {epoch + 1} epochs. Best Validation Loss: {best_loss:.4f}")
            break
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import numpy as np
from collections import Counter

def evaluate_model_classification(model, dataloader):
    model.to('cpu')
    
    all_targets = []
    all_predictions = []

    with torch.no_grad():
        for batch in dataloader:
            inputs, targets = batch
            predictions = model(inputs)

            # Convert predictions to class labels
            if predictions.size(1) > 1:  # Multi-class classification
                predicted_labels = torch.argmax(predictions, dim=1)
            else:  # Binary classification (sigmoid output)
                predicted_labels = (predictions.squeeze() > 0.5).long()

            all_targets.extend(targets.numpy())
            all_predictions.extend(predicted_labels.numpy())

    all_targets = np.array(all_targets)
    all_predictions = np.array(all_predictions)

    accuracy = accuracy_score(all_targets, all_predictions)
    precision = precision_score(all_targets, all_predictions, average="weighted")  # Use "weighted" for multi-class
    recall = recall_score(all_targets, all_predictions, average="weighted")        # Use "weighted" for multi-class
    f1 = f1_score(all_targets, all_predictions, average="weighted")                # Use "weighted" for multi-class
    
    print(Counter(all_predictions))
    
    return accuracy, precision, recall, f1
```

```python
import torch
import numpy as np
from tqdm import tqdm
from sklearn.metrics import accuracy_score, f1_score

def permutation_importance_streaming(model, dataloader, metric="accuracy", n_repeats=5):
    """
    Computes feature importance using permutation importance in a memory-efficient way.

    Parameters:
        model (nn.Module): The trained PyTorch model.
        dataloader (DataLoader): DataLoader for the dataset.
        metric (str): Evaluation metric ('accuracy', 'f1', etc.).
        n_repeats (int): Number of times to shuffle each feature.

    Returns:
        feature_importances (np.array): Importance scores for each feature.
    """
    
    model.eval()
    device = torch.device("cpu")
    model.to(device)

    num_features = None
    baseline_correct = 0
    total_samples = 0
    running_f1 = 0
    batch_count = 0

    # Compute baseline score without shuffling
    with torch.no_grad():
        for batch in dataloader:
            inputs, targets = batch
            inputs = inputs.to(device)

            if num_features is None:
                num_features = inputs.shape[2]  # Extract feature dimension

            outputs = model(inputs).detach().cpu().numpy()
            predictions = np.argmax(outputs, axis=1)
            targets = targets.numpy()

            if metric == "accuracy":
                baseline_correct += (predictions == targets).sum()
                total_samples += len(targets)
            elif metric == "f1":
                batch_f1 = f1_score(targets, predictions, average="weighted")
                running_f1 += batch_f1
                batch_count += 1

    baseline_score = (baseline_correct / total_samples) if metric == "accuracy" else (running_f1 / batch_count)

    # Initialize importance scores
    feature_importances = np.zeros(num_features)

    # Compute permutation importance
    for feature_idx in tqdm(range(num_features), desc="Computing Permutation Importance"):
        permuted_correct = 0
        permuted_f1 = 0
        total_samples = 0
        batch_count = 0

        for _ in range(n_repeats):
            with torch.no_grad():
                for batch in dataloader:
                    inputs, targets = batch
                    inputs = inputs.to(device)

                    # Permute only one feature column
                    inputs_shuffled = inputs.clone()
                    permuted_values = inputs_shuffled[:, :, feature_idx].flatten().numpy()
                    np.random.shuffle(permuted_values)
                    inputs_shuffled[:, :, feature_idx] = torch.tensor(permuted_values.reshape(inputs.shape[0], inputs.shape[1]))

                    # Get model predictions on permuted data
                    outputs = model(inputs_shuffled).detach().cpu().numpy()
                    predictions = np.argmax(outputs, axis=1)
                    targets = targets.numpy()

                    permuted_correct += (predictions == targets).sum()
                    total_samples += len(targets)

        # Compute mean permuted score
        permuted_score = permuted_correct / total_samples
        feature_importances[feature_idx] = baseline_score - permuted_score  # Importance = drop in performance

    return feature_importances

# Run feature importance calculation
feature_importances = permutation_importance_streaming(model, test_dataloader, metric="accuracy", n_repeats=1)
```

```python
def transform_data_tft(
        df: pd.DataFrame,
        with_sliding_window: bool = False,
        size: int = 256,
        step: int = 128,
        classification: bool = True
) -> Tuple[np.ndarray, np.ndarray]:
    """
    Prepares data for TemporalFusionTransformer by converting the DataFrame
    into the required format for TFT.

    :param df: dataframe containing time-series data
    :param with_sliding_window: flag to indicate if sliding window should be used
    :param size: size of sliding window
    :param step: step of sliding window
    :param classification: whether the task is classification or regression
    :return: tuple consisting of sequences and targets in TFT format
    """
    df = df.sort_values(by=['hadm_id', 'charttime'])
    df = df.drop(columns=['subject_id'])

    # Preprocess chartevent_tests
    for test in self.chartevent_tests:
        if 'value' in test:
            if df[test].dtype in (int, float):
                df[[test]] = self.scaler.fit_transform(df[[test]])
                df[[test]].interpolate(method='linear', inplace=True)
            else:
                df[test].ffill(inplace=True)
                encoder = LabelEncoder()
                df[test] = encoder.fit_transform(df[test].astype(str))
        elif 'warning' in test:
            df[test] = df[test].fillna(0)
            df[test] = df[test].astype(int)

    df.fillna(value=-1, inplace=True)

    sequences = []
    los_targets = []
    group_ids = []  # This will hold the group_id for each time series (hadm_id)
    time_idx = []  # This will hold the time indices for each time series

    # Iterate through each patient (group of data)
    for hadm_id, group in df.groupby('hadm_id'):
        # Fetch final time (death or discharge time)
        final_time = self.dt.fetch_data(
            f"""
            SELECT COALESCE(i.outtime, a.deathtime) 
            FROM raw.icustays i
            JOIN raw.admissions a
            ON i.hadm_id = a.hadm_id
            WHERE i.hadm_id = {hadm_id}
            """
        ).iloc[0].tolist()[0]

        sequence = group[self.chartevent_tests].values

        # Sliding window
        if with_sliding_window:
            for start in range(0, len(sequence) - size + 1, step):
                end = start + size
                window = sequence[start:end]
                remaining_los = (final_time - group['charttime'].iloc[end - 1]).total_seconds() / 3600 / 24
                sequences.append(window)
                group_ids.append(hadm_id)
                time_idx.append(np.arange(start, end))  # Time indices for each sequence
                if classification:
                    remaining_los = classify_los(remaining_los)
                los_targets.append(remaining_los)
        else:
            los = (final_time - group['charttime'].min()).total_seconds() / 3600 / 24
            if classification:
                los = classify_los(los)
            sequences.append(sequence)
            los_targets.append(los)
            group_ids.append(hadm_id)
            time_idx.append(np.arange(len(sequence)))  # Time indices for each sequence

    # Convert lists to numpy arrays (without padding)
    sequences = np.array(sequences)
    los_targets = np.array(los_targets)
    group_ids = np.array(group_ids)
    time_idx = np.array(time_idx)

    # Return sequences and targets for TFT without padding
    return sequences, los_targets, group_ids, time_idx
```

```python
import torch
from pytorch_forecasting import Baseline, TemporalFusionTransformer, TimeSeriesDataSet
from torch.utils.data import Dataset, DataLoader
from transformers import Trainer, TrainingArguments
from pathlib import Path
import pickle
import numpy as np


class TimeSeriesHFDataset(Dataset):
    def __init__(self, sequences, targets):
        """
        sequences: numpy array of shape (num_samples, time_steps, num_features)
        targets: numpy array of shape (num_samples,)
        """
        self.sequences = torch.tensor(sequences, dtype=torch.float32)
        self.targets = torch.tensor(targets, dtype=torch.float32)

    def __len__(self):
        return len(self.sequences)

    def __getitem__(self, idx):
        return {
            "inputs_embeds": self.sequences[idx],
            "labels": self.targets[idx]
        }


def load_pretrained_tft(pretrained_model_dir):
    """
    Load the pre-trained Temporal Fusion Transformer model.

    Args:
        pretrained_model_dir: Path to the directory with pre-trained TFT weights and config.

    Returns:
        Loaded Temporal Fusion Transformer model.
    """
    checkpoint = torch.load(pretrained_model_dir + "/model/tft_check_q.ckpt", map_location=torch.device("cuda"))
    hyper_parameters = checkpoint["hyper_parameters"]
    model = TemporalFusionTransformer(**hyper_parameters)
    
    return model


def fine_tune_tft(pretrained_model_dir, train_dataloader, val_dataloader):
    model = load_pretrained_tft(pretrained_model_dir)

    training_args = TrainingArguments(
        output_dir="./results",
        evaluation_strategy="epoch",
        save_strategy="epoch",
        learning_rate=1e-4,
        per_device_train_batch_size=8,
        num_train_epochs=10,
        save_total_limit=2,
        logging_dir="./logs",
        logging_steps=10,
        load_best_model_at_end=True,
        metric_for_best_model="eval_loss",
        no_cuda=False,
    )

    trainer = Trainer(
        model=model,
        args=training_args,
        train_dataset=train_dataloader.dataset,
        eval_dataset=val_dataloader.dataset,
    )

    trainer.train()





pretrained_model_dir = "./TemporalFusionTransformer"
train_dataset = TimeSeriesHFDataset(X_train, y_train)
train_dataloader = DataLoader(train_dataset, batch_size=8, shuffle=True)

val_dataset = TimeSeriesHFDataset(X_val, y_val)
val_dataloader = DataLoader(val_dataset, batch_size=8, shuffle=True)

fine_tune_tft(pretrained_model_dir, train_dataloader, val_dataloader)
```
